```python
# Import packages
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Read the rod simulation data
aktivitaeten = []

for index, file in enumerate(os.listdir("./data_processed")):
    start = file.find("et")
    end = file.find(".npy")
    aktivitaeten.append(file[start+2:end])

aktivitaeten.sort()
aktivitaeten = aktivitaeten[::3]

amount_of_data =len(aktivitaeten)

data = np.zeros([amount_of_data*1000, 128*128]).astype('int8')

for index, file in enumerate(aktivitaeten):
    data[index*1000:(index+1)*1000, 0:128*128] = np.load("./data_processed/aktivitaet"+ file + ".npy")
    print(index)
```

```python
# Separate between the phases and calculate the macroscopic observables
aktivitaeten = np.array(aktivitaeten, dtype ='int')
aktivitaeten_unordered_amount = len(np.where(aktivitaeten < 1.82 * 10**(12))[0])
aktivitaeten_ordered_amount = len(np.where(aktivitaeten > 2.02 * 10**(12))[0])

index = np.arange(amount_of_data*1000)

mask_ordered = index >= (amount_of_data-aktivitaeten_ordered_amount)*1000
mask_unordered = index < aktivitaeten_unordered_amount*1000
mask_critical = [not (a or b) for a, b in zip(mask_ordered, mask_unordered)]

density = np.zeros(amount_of_data*1000)
N_h = np.zeros(amount_of_data*1000)
N_v = np.zeros(amount_of_data*1000)
for n in range(amount_of_data*1000):
    density[n]=len(np.where(data[n,:] != 0)[0])/(128*128)
    N_h[n] = len(np.where(data[n,:] ==10)[0])
    N_v[n] = len(np.where(data[n,:] ==20)[0])

N = N_h + N_v
S = (N_h - N_v)
S_norm = (N_h-N_v)/N
```

```python
# Change representation of the data for PCA
for n in range(amount_of_data*1000):
    data[n, np.where(data[n,:] == 10)[0]] = 1
    for horizontal in range(11,20):
        data[n, np.where(data[n,:] == horizontal)[0]] = 1
    data[n, np.where(data[n,:] == 20)[0]] = -1
    for vertical in range(21,30):
        data[n, np.where(data[n,:] == vertical)[0]] = -1
```

```python
# Perform PCA
pca = PCA(n_components=50, copy=False)
PCC = pca.fit_transform(data)
print(pca.explained_variance_ratio_)  
```

```python
# Plot the first two principal components
plt.plot(PCC[mask_critical,0], PCC[mask_critical,1], 'go', markersize=2)
plt.plot(PCC[mask_ordered,0], PCC[mask_ordered,1], 'ro', markersize=2)
plt.plot(PCC[mask_unordered,0], PCC[mask_unordered,1], 'bo', markersize=2)
plt.legend(['critical' , 'ordered', 'unordered'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.savefig('Rod_PC1PC2.pdf')
```

```python
# Plot the first principal component against the density
plt.plot(density[mask_critical], -PCC[mask_critical,0], 'gx', markersize=2)
plt.plot(density[mask_ordered], -PCC[mask_ordered,0], 'rx', markersize=2)
plt.plot(density[mask_unordered], -PCC[mask_unordered,0], 'bx', markersize=2)
plt.legend(['critical' , 'ordered', 'unordered'])
plt.xlabel(r'density $\rho$')
plt.ylabel('PC1')
plt.savefig('Rod_PC1density.pdf')
```

```python
# Plot the first principal component against the normalized and unnormalized order parameter S
plt.figure(figsize=(11,6)) 
plt.subplots_adjust(wspace=0.4,hspace=0.5)
plt.subplot(121)

plt.plot(S_norm[mask_critical], PCC[mask_critical,0], 'go', markersize=2)
plt.plot(S_norm[mask_ordered], PCC[mask_ordered,0], 'ro', markersize=2)
plt.plot(S_norm[mask_unordered], PCC[mask_unordered,0], 'bo', markersize=2)
plt.legend(['critical' , 'ordered', 'unordered'])
plt.xlabel(r'order parameter $\bar{S}$')
plt.ylabel('PC1')

plt.subplot(122)

plt.plot(S[mask_critical], PCC[mask_critical,0], 'go', markersize=2)
plt.plot(S[mask_ordered], PCC[mask_ordered,0], 'ro', markersize=2)
plt.plot(S[mask_unordered], PCC[mask_unordered,0], 'bo', markersize=2)
plt.legend(['critical' , 'ordered', 'unordered'])
plt.xlabel('unnormalized order parameter S')
plt.ylabel('PC1')
plt.savefig('Rod_PC1S.pdf')
```

```python
# Plot the explained variance ratio
print(pca.explained_variance_ratio_)  

plt.semilogy(np.arange(1,51), pca.explained_variance_ratio_, 'bx', markersize = 5)
plt.xlabel('principal component number')
plt.ylabel('Explained variance ratio')
plt.xticks(np.array([1,5,10,20,30,40,50]), [1,5,10,20,30,40,50])
plt.savefig('ExplainedVar_Rod.pdf')
```

```python
# Plot the absolute value of the order parameter S against the density times rod length
density_av = np.zeros(10)
S_av = np.zeros(10)
for i in range(10):
    density_av[i] = np.mean(density[i*2000:(i+1)*2000])
    S_av[i] = np.mean(S[i*2000:(i+1)*2000])

plt.plot(density_av*8, S_av, 'rx')

plt.xlabel('density * L')
plt.ylabel('Abs order parameter |S|')
```
